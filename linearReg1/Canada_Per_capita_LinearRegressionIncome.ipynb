```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Read the dataset
df=pd.read_csv("canada_per_capita_income.csv")
df.head() #head of the dataset
```

```python
# relation bw year and per capita income (US$) columns of the dataset before the prediction 
%matplotlib inline
plt.xlabel('year', fontsize=20)
plt.ylabel('income in $', fontsize=20)
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
#plt.plot(df.Area, reg.predict(df[['Area']]),color='blue')
```

```python
# initialize the linear regression model
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)']) # fit the model
```

```python
# Predict the model
reg.predict([[2020]])
```

### Verfication of the Linear Regression machine learning algorithm with the help of y = m * x + b

```python
# m= coefficient
m=reg.coef_
```

```python
# c=intercept
c= reg.intercept_
```

```python
# y is the dependent variable, 
#y=mx + c
y=m*2020 + c
y
#135.78767123 * 3300+ 180616.43835616432 
```
